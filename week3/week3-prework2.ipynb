## 【課題5】用語の調査
下記の用語について調査してまとめる。
- 尖度 (参考: https://bellcurve.jp/statistics/glossary/2113.html)

分布が正規分布からどれだけ尖っているかを表す統計量である。山の尖り度と裾の広がり度を反映している。正規分布よりも尖った分布で裾が重いときには正の値を、正規分布より扁平な分布で裾が軽い時は負の値をとる。正規分布の場合は値が0となる。
下記の式で定義されている。
$$ \frac{n(n + 1)}{(n - 1)(n - 2)(n - 3)}\sum_{i = 1}^{n}\frac{(x_i - \overline{x})^4}{s^4} - \frac{3(n - 1)^2}{(n - 2)(n - 3)} $$
<br>
- 歪度 (参考: https://bellcurve.jp/statistics/glossary/2135.html)

分布が正規分布からどれだけ歪んでいるかを表す統計量である。左右対称性の度合いを反映している。左に偏った（右裾が長い）分布のときには正の値を、右に偏った（左の裾が長い）分布のときには負の値をとる。正規分布のような左右対象の場合には0になる。
下記の式で定義されている。
$$ \frac{n}{(n - 1)(n - 2)}\sum_{i=1}^{n}\frac{(x_i - \overline{x})^3}{s^3} $$

## 【課題6】分布の確認
- seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。
- 目的変数に対して対数変換を行う。
- 対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。

```python
#目的変数のdistribution plotを表示し、尖度と歪度も算出する
import seaborn as sns
sns.set()
sns.distplot(data.SalePrice);
```

```python
print('kurtosis: {:.3f}'.format(data.SalePrice.kurt()))
print('skewness: {:.3f}'.format(data.SalePrice.skew()))
```

```python
#目的変数に対して対数変換を行う
data['SalePrice_log'] = np.log(data.SalePrice)
```

```python
#対数変換した目的変数に対してdistribution plotを作成し、尖度と歪度を算出する
sns.distplot(data.SalePrice_log);
```

```python
print('kurtosis: {:.3f}'.format(data.SalePrice_log.kurt()))
print('skewness: {:.3f}'.format(data.SalePrice_log.skew()))
```

尖度(kurtosis)、歪度(skewness)は、対数変換を行うことで共に小さくなった。  
対数変換前: {Skewness: 6.529, Kurtosis: 1.882}, 対数変換後: {Skewness: 0.807, Kurtosis: 0.121}  
正規分布は尖度、歪度0の分布であるから、対数変換によって目的関数の分布の形状が正規分布に近づいたと言える。

## 【課題7】相関係数の確認
特徴量の確認をしていきます。それぞれに対してマークダウンで説明や考察を書いてください。
- 相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。
- 目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。
- 選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。
- 選んだ10個の特徴量の中でお互いの相関係数が高い組み合わせを3つ探し出す。

**補足**  
重回帰分析などで予測する場合、相関が高すぎる特徴量の組み合わせがあると予測に悪影響が出ることがあります。  
このような場合の簡単な対処方法として、そういった特徴量の組みを探し出して、１つに絞るということがあります。

```python
#相関行列のヒートマップを作る
data.corr()
```

```python
sns.heatmap(data.corr());
```

```python
#目的変数と相関の高い特徴量10個のヒートマップを作る
#SalePriceとSalePrice_logを除くため2:12でスライス
high_corr_10 = data.corr().SalePrice_log.sort_values(ascending=False)[2:12]
print('目的関数相関の高い特徴量')
print(high_corr_10)
```

```python
sns.heatmap(data[high_corr_10.index].corr());
```

#### 目的関数と相関の高い特徴量の詳細
OverallQual: 家全体の材質や仕上げに対する品質  
GrLivArea: 地上にある居住空間の広さ(feet\**2)  
GarageCars: ガレージに入る車の数  
GarageArea: ガレージの広さ(feet\**2)  
TotalBsmtSF: 地階の広さ(feet\**2)  
1stFlrSF: 1階の広さ(feet\**2)  
FullBath: 地上にあるお風呂場の数  
TotRmsAbvGrd: 地上にある部屋数の数（お風呂場は除く）  
Yearbuilt: 建築された年  
YearRemodAdd: 改築された年  

```python
#相関係数の高い特徴量の組み合わせ
data[high_corr_10.index].corr().unstack().sort_values(ascending=False)[:20]
```

相関係数が高い組み合わせは、  
\(GarageCars, GarageArea\)  
\(TotRmsAbvGrd, GrLivArea\)   
\(TotalBsmtSF, 1stFlrSF\)
