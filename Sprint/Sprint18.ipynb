```python
def mytestGenerator(test_path,num_image = 30,target_size = (256,256),flag_multi_class = False,as_gray = True):
    
    test_image_list = [name for name in os.listdir(test_path) if not name.startswith('.')]

    for i in test_image_list:
        img = io.imread(test_path+'/'+i, as_gray = as_gray)
        img = img / 255
        img = trans.resize(img,target_size)
        img = np.reshape(img,img.shape+(1,)) if (not flag_multi_class) else img
        img = np.reshape(img,(1,)+img.shape)
        yield img
```

```python
testGene = mytestGenerator('/content/drive/My Drive/Salt/test_data_resize/images')
model = unet()
model.load_weights("salt.hdf5")
results = model.predict_generator(testGene,30,verbose=1)
saveResult('/content/drive/My Drive/Salt',results)
```
