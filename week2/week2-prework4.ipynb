```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 【問題1】ダーツのまとを選ぶ

```python
select_target = lambda: np.random.choice(np.arange(1,7))
select_target()
```

## 【問題2】ダーツを投げて、刺さった場所を記録する

```python
def get_xy(target):
    x = np.random.normal(0, 0.1) + target
    y = np.random.normal(0, 0.1)
    return x, y
```

## 【問題3】ダーツをたくさん投げて、まとの番号と紐付けた状態で記録する
ダーツを1000回投げて、全てのダーツの刺さった位置をディクショナリに格納してください。  
ディクショナリのkeyはダーツのまとの番号になります。

```python
result_dict = {}
for _ in range(1000):
    target = select_target()
    x, y = get_xy(target)
    result_dict[target] = result_dict.get(target,[]) + [[x,y]]
```

## 【問題4】ダーツのまとと刺さった位置を描画する
問題3で作成したディクショナリからkeyごとに値を取り出し、まとごとに刺さった場所を描画してみましょう。  
ダーツのまとを表現するためここではmatplotlibのpatchesクラスを用いてください。

```python
from matplotlib import patches
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 4))
for target, coord in result_dict.items():
    coord = np.array(coord)
    plt.scatter(coord[:,0], coord[:,1])
    ax.add_patch(patches.Circle(xy=(target,0), radius=0.1, ec='b', fc='None'))
    ax.add_patch(patches.Circle(xy=(target,0), radius=0.3, ec='b', fc='None'))
    ax.add_patch(patches.Circle(xy=(target,0), radius=0.5, ec='b', fc='None'))
plt.axis('scaled');
```

## 【問題5】関数化
ダーツを投げる回数、ダーツのまとの数が変わっても描画できるように、問題1~問題4で作成したコードを関数化しましょう。

```python
def darts(n_throw, n_target):
    result_dict = {}
    
    for _ in range(n_throw):
        target = np.random.choice(np.arange(1,n_target+1))
        x = np.random.normal(0, 0.1) + target
        y = np.random.normal(0, 0.1)
        result_dict[target] = result_dict.get(target,[]) + [[x,y]]
    
    fig, ax = plt.subplots(1, 1, figsize=(16, 4))
    for target, coord in result_dict.items():
        coord = np.array(coord)
        plt.scatter(coord[:,0], coord[:,1])
        ax.add_patch(patches.Circle(xy=(target,0), radius=0.1, ec='b', fc='None'))
        ax.add_patch(patches.Circle(xy=(target,0), radius=0.3, ec='b', fc='None'))
        ax.add_patch(patches.Circle(xy=(target,0), radius=0.5, ec='b', fc='None'))
    plt.axis('scaled');
```

```python
darts(1000, 6)
```
